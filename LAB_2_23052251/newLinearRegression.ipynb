```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error ,r2_score
import matplotlib.pyplot as plt

house_data = pd.read_csv("boston.csv")

#indepent
x = house_data['RM'].values.reshape(-1,1) # .values for convert into 1-d array
# .reshape is used for convert them into 2-D where -1 is for row and 1 for col

#dependent
y = house_data["MEDV"].values

#Split data 
x_train,x_test,y_train, y_test = train_test_split(x, y , test_size = 0.3, random_state = 42)

print(f"Training data size: {x_train.shape[0]} samples")
print(f"Testing data size: {x_test.shape[0]} samples")


# 2 Model Training
train_model = LinearRegression()

train_model.fit(x_train, y_train)

y_pred = train_model.predict(x_test)

#3 Model_Coeeficients y=mx + c
# y is predicted value 
# x is feature used for prediction
# C is intercept( It is the starting point of line when it is zero)
# m meansure the direction of line

print(f"\nModel Intercept c: {train_model.intercept_:.2f}")
print(f"Model Coefficient (m): {train_model.coef_[0]:.2f}")

# 4 model Evulation
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("\nModel Evaluation Metrics:")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R-squared (R2 Score): {r2:.4f}")

#Model_Visulization
plt.figure(figsize=(10, 6))

# 1. Scatter plot of the actual test data
plt.scatter(x_test, y_test, color='blue', label='Actual Prices (Test Data)')

#2 Plot the regression Line
plt.plot(x_test, y_pred, color='red', linewidth=3, label=f'Regression Line (R2: {r2:.2f})')

plt.title('Simple Linear Regression: House Price (MEDV) vs. Rooms (RM)')
plt.xlabel('Average Number of Rooms (RM)')
plt.ylabel('Median House Price (MEDV) in $1000s')
plt.legend()
plt.grid(True)
plt.show()

















```

```python
#Multi Linear Regression

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Load the data (assuming "boston.csv" is still available)
data = pd.read_csv("boston.csv")


# We select 'RM', 'LSTAT', and 'PTRATIO' as independent variables
features = ['RM', 'LSTAT', 'PTRATIO']
X = data[features]  # Independent variables (3 columns)
y = data['MEDV']    # Dependent variable (House Price)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8, random_state=42)

print(f"Features used for prediction: {features}")
print(f"Training data shape: {X_train.shape}")

# Create and train the Multiple Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

print("\n### Model Coefficients (w) ###")
for feature, coef in zip(features, model.coef_):
    print(f"{feature:<7} Coefficient: {coef:9.3f}")

print(f"Intercept (w0): {model.intercept_:15.3f}")
# Calculate evaluation metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("\n### Model Evaluation Metrics ###")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R-squared (R2 Score): {r2:.4f} (Likely improved from single-feature model)")
# Visualization of Actual vs. Predicted Values
plt.figure(figsize=(8, 8))
plt.scatter(y_test, y_pred, color='purple', alpha=0.6)

# Plot the ideal line (where Actual = Predicted)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--', lw=2, label='Perfect Prediction Line')

plt.title('Actual Prices vs. Predicted Prices (Multiple Regression)')
plt.xlabel('Actual MEDV (Test Set)')
plt.ylabel('Predicted MEDV')
plt.legend()
plt.grid(True)
plt.show()

#
```
