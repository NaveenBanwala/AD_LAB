```python
from tensorflow.keras.models import load_model

# Load your model
model = load_model('my_model.keras')
```

```python
from tensorflow.keras.preprocessing import image
import numpy as np

# Path to your test image
img_path = 'mehaki.png'

# Load the image and resize it to match model input
# Example: if your model expects 28x28 grayscale images
img = image.load_img(img_path, target_size=(28, 28), color_mode='grayscale')

# Convert to array
img_array = image.img_to_array(img)

# Normalize if you did during training
img_array = img_array / 255.0

# Add batch dimension (model expects shape: (1, 28, 28, 1))
img_array = np.expand_dims(img_array, axis=0)
```

```python
from tensorflow.keras.preprocessing import image
import numpy as np

img_path = 'image.png'

# Get height, width, channels from model
_, h, w, c = model.input_shape

# Load and resize image
if c == 1:
    color_mode = 'grayscale'
else:
    color_mode = 'rgb'

img = image.load_img(img_path, target_size=(h, w), color_mode=color_mode)

# Convert to array
img_array = image.img_to_array(img)

# Normalize if needed
img_array = img_array / 255.0

# Add batch dimension
img_array = np.expand_dims(img_array, axis=0)

# Predict
pred = model.predict(img_array)
predicted_class = np.argmax(pred)
print("Predicted Class:", predicted_class)
```
